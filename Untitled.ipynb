```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.utils.class_weight import compute_class_weight
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

```python
# Step 1: Load the Dataset
def load_datasets(training_path, testing_path):
    train_df = pd.read_csv(training_path)
    test_df = pd.read_csv(testing_path)
    print(f"Training dataset loaded with {train_df.shape[0]} rows and {train_df.shape[1]} columns.")
    print(f"Testing dataset loaded with {test_df.shape[0]} rows and {test_df.shape[1]} columns.")
    return train_df, test_df

# File paths (replace with your actual paths)
training_path = r'UNSW_NB15_training-set.csv'
testing_path = r'UNSW_NB15_testing-set.csv'

# Load the datasets
train_dataset, test_dataset = load_datasets(training_path, testing_path)
```

```python
# Importing the necessary libraries for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing the distribution of the target variable (label)
plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=train_dataset)
plt.title('Distribution of Attack vs Normal Traffic')
plt.xlabel('Label (0: Normal, 1: Attack)')
plt.ylabel('Count')
plt.show()
```

```python
# Visualizing the distribution of some key features
features_to_plot = ['dur', 'spkts', 'sbytes', 'rate', 'sload', 'dbytes']

# Plotting histograms for selected features
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(2, 3, i)
    sns.histplot(train_dataset[feature], kde=True, bins=50)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Visualizing the distribution of categorical features
categorical_features = ['proto', 'service', 'state', 'attack_cat']

# Plotting bar plots for selected categorical features
plt.figure(figsize=(12, 10))
for i, feature in enumerate(categorical_features, 1):
    plt.subplot(2, 2, i)
    sns.countplot(x=feature, data=train_dataset, palette='Set2')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Plotting boxplots for numerical features against the target variable (label)
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features_to_plot, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='label', y=feature, data=train_dataset, palette='Set2')
    plt.title(f'{feature} vs Label')
    plt.xlabel('Label (0: Normal, 1: Attack)')
    plt.ylabel(feature)
plt.tight_layout()
plt.show()
```

```python
# Step 2: Feature Selection
def feature_selection(df):
    # Keep selected features
    selected_columns = ['proto', 'service', 'spkts', 'dpkts', 'sbytes', 'dbytes', 'rate', 'dur', 'attack_cat']
    df = df[selected_columns].copy()

    # Use `dur` for rate calculation
    df['rate'] = (df['sbytes'] + df['dbytes']) / df['dur'].replace(0, np.nan)

    # Add derived features
    df['length'] = df['sbytes'] + df['dbytes']
    df['direction'] = df['spkts'] / df['dpkts'].replace(0, np.nan)
    df['size_variance'] = ((df['sbytes'] - df['dbytes']) ** 2).replace(0, np.nan)

    return df

# Apply feature selection to training and testing datasets
train_features = feature_selection(train_dataset)
test_features = feature_selection(test_dataset)
```

```python
# Step 3: Impute Missing Values for Numerical Columns
numerical_columns = ['spkts', 'dpkts', 'sbytes', 'dbytes', 'rate', 'dur', 'length', 'direction', 'size_variance']
categorical_columns = ['proto', 'service', 'attack_cat']

# Impute numerical features
imputer = SimpleImputer(strategy='mean')
train_features[numerical_columns] = imputer.fit_transform(train_features[numerical_columns])
test_features[numerical_columns] = imputer.transform(test_features[numerical_columns])
```

```python
# Step 4: Align Columns
for col in train_features.columns:
    if col not in test_features.columns:
        test_features[col] = 0
test_features = test_features[train_features.columns]
```

```python
# Step 5: Encode Categorical Features
def encode_categorical_features(df):
    return pd.get_dummies(df, columns=['proto', 'service'])

train_encoded = encode_categorical_features(train_features)
test_encoded = encode_categorical_features(test_features)

# Align columns after encoding
for col in train_encoded.columns:
    if col not in test_encoded.columns:
        test_encoded[col] = 0
test_encoded = test_encoded[train_encoded.columns]
```

```python
# Step 6: Normalize Numerical Features
numerical_features = ['spkts', 'dpkts', 'sbytes', 'dbytes', 'rate', 'dur', 'length', 'direction', 'size_variance']
scaler = StandardScaler()

train_encoded[numerical_features] = scaler.fit_transform(train_encoded[numerical_features])
test_encoded[numerical_features] = scaler.transform(test_encoded[numerical_features])

# Save the scaler for real-time use
joblib.dump(scaler, 'scaler.pkl')
print("Scaler updated and saved successfully.")
```

```python
# Step 7: Prepare Features and Target
X_train = train_encoded.drop(columns=['attack_cat'])
y_train = train_encoded['attack_cat']

X_test = test_encoded.drop(columns=['attack_cat'])
y_test = test_encoded['attack_cat']

# Save only the feature columns for real-time processing
feature_columns = X_train.columns
joblib.dump(feature_columns, 'X_train_columns.pkl')
print("Feature columns saved successfully as 'X_train_columns.pkl'.")
```

```python
# Step 8: Balance Dataset Using SMOTE
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

print("Class Distribution After Balancing:")
print(pd.Series(y_train_balanced).value_counts())
```

```python
# Step 9: Encode Target Variable
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_balanced)
y_test_encoded = label_encoder.transform(y_test)

# Save the label encoder
joblib.dump(label_encoder, 'label_encoder.pkl')
print("Label encoder saved successfully.")
```

```python
# Step 10: Train Models
print("Training RandomForestClassifier...")
rf_model = RandomForestClassifier(random_state=42, n_estimators=200, max_depth=15, min_samples_split=5)
rf_model.fit(X_train_balanced, y_train_encoded)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test_encoded, rf_predictions)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")

print("Training DecisionTreeClassifier...")
dt_model = DecisionTreeClassifier(random_state=42, max_depth=10, min_samples_split=5)
dt_model.fit(X_train_balanced, y_train_encoded)
dt_predictions = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test_encoded, dt_predictions)
print(f"Decision Tree Accuracy: {dt_accuracy:.4f}")

print("Training XGBClassifier...")
xgb_model = XGBClassifier(eval_metric='mlogloss', random_state=42, learning_rate=0.1, max_depth=7, n_estimators=100)
xgb_model.fit(X_train_balanced, y_train_encoded)
xgb_predictions = xgb_model.predict(X_test)
xgb_accuracy = accuracy_score(y_test_encoded, xgb_predictions)
print(f"XGBClassifier Accuracy: {xgb_accuracy:.4f}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Evaluate Random Forest Classifier
rf_accuracy = accuracy_score(y_test_encoded, rf_predictions)
print("Random Forest Performance:")
print("Accuracy:", rf_accuracy)
print("\nConfusion Matrix:\n", confusion_matrix(y_test_encoded, rf_predictions))
print("\nClassification Report:\n", classification_report(y_test_encoded, rf_predictions))

# Evaluate Decision Tree Classifier
dt_accuracy = accuracy_score(y_test_encoded, dt_predictions)
print("\nDecision Tree Performance:")
print("Accuracy:", dt_accuracy)
print("\nConfusion Matrix:\n", confusion_matrix(y_test_encoded, dt_predictions))
print("\nClassification Report:\n", classification_report(y_test_encoded, dt_predictions))

# Evaluate XGBoost Classifier
xgb_accuracy = accuracy_score(y_test_encoded, xgb_predictions)
print("\nXGBoost Performance:")
print("Accuracy:", xgb_accuracy)
print("\nConfusion Matrix:\n", confusion_matrix(y_test_encoded, xgb_predictions))
print("\nClassification Report:\n", classification_report(y_test_encoded, xgb_predictions))
```
